Uma plataforma monitora e controla a potência elétrica entregue a veículos em um eletroposto comercial. Suponha que, em uma determinada sessão de recarga controlada, a potência instantânea entrege pelo sistema, em kW, seja modelada por:  
P(t) = 18 + ((24 * t) / (t + 2)),     t >= 0  
em que t representa o tempo, em minutos, contado a partir do início da sessão

#### Import de bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

#### Definição da função

```python
#simbolica (sympy)
#t
t_sp = sp.symbols('t')
#funcao
p_sp = 18 + ((24 * t_sp) / (t_sp + 2))

#numerica (numpy)
#intervalo de t
t = np.linspace(0, 10, 50)
#funcao
def p(t):
    return 18 + ((24 * t) / (t + 2))
```

### a. Gráfico da função

```python
#plot do gráfico
plt.figure()

plt.plot(t, p(t), label="P(t) = 18 + ((24 * t) / (t + 2))", color = "blue")

plt.title("Gráfico da Potência P(t) ao longo do tempo")
plt.xlabel("Tempo (min)")
plt.ylabel("Potência (kW)")

plt.legend()
plt.grid()

plt.show()
```

### b. Limites para t -> 0 e para t -> 8

```python
#calculo dos limites
lim_0 = sp.limit(p_sp, t_sp, 0)
lim_8 = sp.limit(p_sp, t_sp, 8)

#plot do gráfico lim_0
plt.figure()

plt.plot(t, p(t), label="P(t) = P(t) = 18 + ((24 * t) / (t + 2))", color = "blue")
plt.scatter(0, float(lim_0), label="Limite quando t->0",zorder = 2, facecolors = 'white', edgecolors = 'green')
plt.annotate(f"(0, {float(lim_0):.1f})", (0, float(lim_0)), textcoords="offset points", xytext=(10, 8),
             bbox=dict(facecolor="white", edgecolor="white"))

plt.title("Limite da função quando t tende a 0")
plt.xlabel("Tempo (min)")
plt.ylabel("Potência (kW)")

plt.legend()
plt.grid()

plt.show()

#plot do gráfico lim_8
plt.figure()

plt.plot(t, p(t), label="P(t) = P(t) = 18 + ((24 * t) / (t + 2))", color = "blue")
plt.scatter(8, float(lim_8), label="Limite quando t->8",zorder = 2, facecolors = 'white', edgecolors = 'red')
plt.annotate(f"(0, {float(lim_8):.1f})", (8, float(lim_8)), textcoords="offset points", xytext=(8,-15),
            bbox=dict(facecolor="white", edgecolor="white"))

plt.title("Limite da função quando t tende a 8")
plt.xlabel("Tempo (min)")
plt.ylabel("Potência (kW)")

plt.legend()
plt.grid()

plt.show()
```

Os limites encontrados indicam como a potência se comporta em momentos específicos da recarga. O limite quando t -> 0 representa a potência inicial fornecida pelo sistema, logo no início do carregamento, mostrando que a recarga começa com um valor definido.  
Já o valor do limite quando t -> 8  indica a potência fornecida após alguns minutos de recarga, permitindo observar como o sistema evolui ao longo do tempo. Comparando esses valores, percebe-se que a potência aumenta conforme o tempo passa, mas esse crescimento ocorre de forma gradual.  
Isso indica que o sistema de recarga é projetad para fornecer energia de maneira estável e progressiva, garantindo eficiência e segurança durante o processo. 

### c. Variação média de potencia entre t=1 e t=3

fórmula para cálculo de variação média de uma função:  
( f(b) - f(a) ) / ( b - a )

```python
P1 = p(1)
P3 = p(3)

variacao = (P3 - P1) / (3 - 1)

print(f"A variação média da potência, entre 1 e 3 minutos, é igual a: {variacao:.1f} kW por minuto")
```

### d. Derivada analítica de P(t)

```python
#calculo da derivada
p_diff = sp.diff(p_sp, t_sp)

#mostrar
print("A derivada P'(t) de P(t) é:")
display(p_diff)
print("Ou, de forma simplificada:")
display(sp.simplify(p_diff))
```

#### e. Derivadas de P para t = 1 e t = 8 (P'(1) e P'(8))

```python
#mostrar
print(f"P'(1): {p_diff.subs(t_sp, 1):.2f}")
print(f"P'(8): {p_diff.subs(t_sp, 8):.2f}")
```

#### f. Como a derivada P'(t) se relaciona com o comportamento do gráfico de P(t)

Numericamente, a derivada de uma função representa, em um determinado ponto, a taxa de variação instantânea da função. Geometricamente, indica o coeficiente angular da reta tangente a curva que passa pelo ponto. Em outras palavras, ela indica o quão rápido a função muda em um ponto.  
Um valor de derivada mais alto indica um ângulo maior (maior inclinação) e, consequentemente, mais variação. Uma derivada negativa indica que a função é decrescente. 
No gráfico de P(t), no ponto em que t é igual a 1, ocorre um crescimento acelerado. Enquanto no ponto em que t é igual a 8, o crescimento é mais devagar, tendendo a 0. Como a derivada é sempre positiva, a função nunca decresce.

Portanto, quanto mais preciso o modelo, maior a capacidade do sistema de operar de forma inteligente, segura e eficiente.
